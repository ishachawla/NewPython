# Patterns in Temperatures in Coastal Regions, Continents and totally opposite geographical areas

```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline
from bokeh.charts import Line, show, output_notebook, defaults
from bokeh.plotting import figure, output_file, show
import os
```

```python
a=os.getcwd()
path=a+"\\RawData\\GlobalLandTemperaturesByCountry.csv"
print(path)
```

```python
countrytemp=pd.read_csv(path, encoding='utf8',parse_dates=['dt'],infer_datetime_format=True)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
```

```python
countrytemp=countrytemp.groupby(['Year','Country'])[['AverageTemperature']].mean().reset_index()
```

```python
United=countrytemp[(countrytemp.Country =='United States')] #COnsidering United States Data
```

```python
Brazil=countrytemp[(countrytemp.Country =='Brazil')] #taking into consideration Brazil
```

```python
United=United[-200:]
Brazil=Brazil[-200:]
```

```python
minyear=United['Year'].min()
maxyear=United['Year'].max()
```

```python
sb.set(style="white")
p=sb.color_palette("cubehelix", 8)
p.as_hex()
#sb.set_color_codes("GnBu",10)
ax=plt.subplots(figsize=(8, 6))

plt.plot(United['Year'], United['AverageTemperature'], color='#2ecc71',label='United States')
plt.plot(Brazil['Year'], Brazil['AverageTemperature'], color='blue',label='Brazil')
#ax.fill_between(globaltemp['Year'], all_data['minAvgTemp'], all_data['maxAvgTemp'], color='b')

plt.xlim(minyear, maxyear)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Average Temperature', fontsize=18)
plt.title('Temp. difference over demographics', fontsize=20,color='Grey')
#plt.set_xticklabels(years, rotation=5)
#ax.set_title('Average global temperature including uncertainty')
plt.legend()  
plt.show()
#so we see that over the course of some 100 years Brazil and America move parallel to each other and never intersect.
```

# Considering countries in different continents to see the temperature trends that they follow


```python
Japan=countrytemp[(countrytemp.Country =='Japan')]
Argentina=countrytemp[(countrytemp.Country =='Argentina')]
Canada=countrytemp[(countrytemp.Country =='Canada')]
Ireland=countrytemp[(countrytemp.Country =='Ireland')]
Chad=countrytemp[(countrytemp.Country =='Chad')]
Australia=countrytemp[(countrytemp.Country =='Australia')]
Denmark=countrytemp[(countrytemp.Country =='Denmark')]
Greenland=countrytemp[(countrytemp.Country =='Greenland')]
Russia=countrytemp[(countrytemp.Country =='Russia')]
```

```python
Japan=Japan[-200:]
Argentina=Argentina[-200:]
Canada=Canada[-200:]
Ireland=Ireland[-200:]
Chad=Chad[-200:]
Australia=Australia[-200:]
Denmark=Denmark[-200:]
Greenland=Greenland[-200:]
Russia=Russia[-200:]
```

```python
x=Japan['Year']
y=Japan['AverageTemperature']
```

```python
output_file("multiple.html")
p = figure(plot_width=700, plot_height=400)
p.line(x, y, line_width=5,legend='Japan/ASIA')
p.line(x, Argentina['AverageTemperature'], line_width=5,color='Red',legend='Argentina/SOUTH AMERICA')
p.line(x, Canada['AverageTemperature'], line_width=5,color='Yellow',legend='Canada/NORTH AMERICA')
p.line(x, Ireland['AverageTemperature'], line_width=5,color='Pink',legend='Ireland/EUROPE')
p.line(x, Chad['AverageTemperature'], line_width=5,color='Green',legend='Chad/AFRICA')
p.line(x, Australia['AverageTemperature'], line_width=5,color='Aqua',legend='Australia/AUSTRALIA')
show(p)
#after plotting this we see that Canada is the coldest amongst all and its temperatures do not intersect with anyone 
#similarly Chad is the warmest of all having no trends matching with anyone
#Japan,Argentina and Ireland lie quite close to each other, merely touching at a couple of points.
```

```python
#plotting a graph between United States and Canada since they are quite cold and see if they have any intersecting temperatures
sb.set(style="white")
p=sb.color_palette("cubehelix", 8)
p.as_hex()
#sb.set_color_codes("GnBu",10)
ax=plt.subplots(figsize=(8, 6))

plt.plot(United['Year'], United['AverageTemperature'], color='Red',label='United States')
plt.plot(Canada['Year'], Canada['AverageTemperature'], color='Aqua',label='Canada')
#ax.fill_between(globaltemp['Year'], all_data['minAvgTemp'], all_data['maxAvgTemp'], color='b')

plt.xlim(minyear, maxyear)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Average Temperature', fontsize=18)
plt.title('Temp. difference b/w US and Canada', fontsize=20,color='Grey')
#plt.set_xticklabels(years, rotation=5)
#ax.set_title('Average global temperature including uncertainty')
plt.legend()  
plt.show()
#So Canada is quite cold when compared to United States
```
