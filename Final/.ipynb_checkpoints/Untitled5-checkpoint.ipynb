```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline
from bokeh.charts import Line, show, output_notebook, defaults
from bokeh.plotting import figure, output_file, show
import os
from subprocess import check_output
import sys
```

```python
#sys.path.append(os.getcwd() + 'FinalProject/RawData'
a=os.getcwd()
path=a+"\\RawData\\GlobalLandTemperaturesByCountry.csv"
print(path)
```

```python
countrytemp=pd.read_csv(path, encoding='utf8',parse_dates=['dt'],infer_datetime_format=True)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year) #adding year column
countrytemp['Month'] = countrytemp.dt.apply(lambda x: x.month)
```

```python
countrytemp=countrytemp.groupby(['Year','Country'])[['AverageTemperature','AverageTemperatureUncertainty']].mean().reset_index()
```

```python
countrytemp.sort(['AverageTemperature', 'Year'], ascending=[False, False], inplace=True) #Sorting the dataframe
```

```python
mean=[]
countries= np.unique(countrytemp['Country'])
for Country in countries:
     mean.append(countrytemp[countrytemp['Country'] == 
                                               Country]['AverageTemperature'].mean())
```

```python
#Visual Representation of Hottest to Coldest Nations
mean_temp_bar, countries_bar = (list(x) for x in zip(*sorted(zip(mean, countries), 
                                                             reverse = True)))
sb.set(font_scale=0.9) 
f, ax = plt.subplots(figsize=(4.5, 50))
colors_cw = sb.color_palette('coolwarm', len(countries))
sb.barplot(mean_temp_bar, countries_bar)
Text = ax.set(xlabel='Average temperature', title='Average temperature in countries')
```
