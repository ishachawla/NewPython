```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
countrytemp['Month'] = countrytemp.dt.apply(lambda x: x.month)
citytemp['Year'] = citytemp.dt.apply(lambda x: x.year)
citytemp['Month'] = citytemp.dt.apply(lambda x: x.month)
majorcity['Year'] = majorcity.dt.apply(lambda x: x.month)
majorcity['Month'] = majorcity.dt.apply(lambda x: x.month)
statetemp['Year'] = statetemp.dt.apply(lambda x: x.month)
statetemp['Month'] = statetemp.dt.apply(lambda x: x.month)
globaltemp['Year'] = globaltemp.dt.apply(lambda x: x.month)
globaltemp['Month'] = globaltemp.dt.apply(lambda x: x.month)
```

```python
def define_season(Month):
    if (Month<=6) & (Month>=5): #6,5
        return 'SPRING'
    elif (Month<=4) & (Month>=1): #1,2,3,4
        return 'WINTER'
    elif (Month>=7) & (Month<=9): #7,8,9
        return 'SUMMER'
    elif (Month<=11) & (Month>=10): #10,11
        return 'AUTUMN'
    else:
        return 'WINTER' #12
```

```python
countrytemp['Season']=countrytemp['Month'].apply(define_season)
citytemp['Season']=countrytemp['Month'].apply(define_season)
statetemp['Season']=countrytemp['Month'].apply(define_season)
globaltemp['Season']=countrytemp['Month'].apply(define_season)
majorcity['Season']=countrytemp['Month'].apply(define_season)
```
