# Temperature patterns in the United States based on Seasons

```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
import os
```

```python
a=os.getcwd()
path=a+"\\RawData\\GlobalLandTemperaturesByCountry.csv"
print(path)
```

```python
countrytemp=pd.read_csv(path, encoding='utf8',parse_dates=['dt'],infer_datetime_format=True)
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
countrytemp['Month'] = countrytemp.dt.apply(lambda x: x.month)
```

```python
def define_season(Month):
    if (Month<=6) & (Month>=5): #6,5
        return 'SPRING'
    elif (Month<=4) & (Month>=1): #1,2,3,4
        return 'WINTER'
    elif (Month>=7) & (Month<=9): #7,8,9
        return 'SUMMER'
    elif (Month<=11) & (Month>=10): #10,11
        return 'AUTUMN'
    else:
        return 'WINTER' #12
```

```python
countrytemp['Season']=countrytemp['Month'].apply(define_season)

```

```python
countrytemp=countrytemp.fillna(0)
minyear=countrytemp['Year'].min()
maxyear=countrytemp['Year'].max()
```

```python
countrytemp=countrytemp[(countrytemp.Country =='United States')]
```

```python
countrytemp=countrytemp.groupby(['Year','Season'])[['AverageTemperature']].mean().reset_index()
```

```python
spring=countrytemp[(countrytemp.Season=='SPRING')]
summer=countrytemp[(countrytemp.Season=='SUMMER')]
autumn=countrytemp[(countrytemp.Season=='AUTUMN')]
winter=countrytemp[(countrytemp.Season=='WINTER')]
spring=spring.dropna()
summer=summer.dropna()
autumn=autumn.dropna()
winter=winter.dropna()
```

```python
summer.plot(kind='scatter', x='Year', y='AverageTemperature', ax=ax1,color='#fff100',marker='o',figsize=(16, 8))
ax1.set_title('Summer',fontsize=12,color='Grey')
spring.plot(kind='scatter', x='Year', y='AverageTemperature', ax=ax2,color='#ffd110',marker="o", figsize=(16, 8))
ax2.set_title('Spring',fontsize=12,color='Grey')
autumn.plot(kind='scatter', x='Year', y='AverageTemperature', ax=ax3,color='#ff9130',marker="o",figsize=(16, 8))
ax3.set_title('Autumn',fontsize=12,color='Grey')
winter.plot(kind='scatter', x='Year', y='AverageTemperature', ax=ax4,color='#ff7140',marker="o", figsize=(16, 8))
ax4.set_title('Winter',fontsize=12,color='Grey')
```
