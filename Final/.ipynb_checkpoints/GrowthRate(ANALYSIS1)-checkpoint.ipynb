# Finding the growth rate in temperature every year from 2000-2013

```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline
from bokeh.charts import Line, show, output_notebook, defaults
from bokeh.plotting import figure, output_file, show
import os
```

```python
a=os.getcwd()
path=a+"\\RawData\\GlobalLandTemperaturesByCountry.csv"
print(path)
```

```python
countrytemp=pd.read_csv(path,parse_dates=['dt'],infer_datetime_format=True)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
```

```python
countrytemp=countrytemp.groupby(['Year','Country'])[['AverageTemperature','AverageTemperatureUncertainty']].mean().reset_index()
countrytemp=countrytemp.dropna()
minyear=countrytemp['Year'].min()
maxyear=countrytemp['Year'].max()
```

```python
sb.set(style="white")
p=sb.color_palette("cubehelix", 8)
p.as_hex()
#sb.set_color_codes("GnBu",10)
ax=plt.subplots(figsize=(20, 10))

plt.plot(countrytemp['Year'], countrytemp['AverageTemperature'], color='#2ecc71')
plt.scatter(countrytemp['Year'], countrytemp['AverageTemperatureUncertainty'],marker='+', color='blue')
#ax.fill_between(globaltemp['Year'], all_data['minAvgTemp'], all_data['maxAvgTemp'], color='b')

plt.xlim(minyear, maxyear)
plt.xlabel('Year', fontsize=18,color='Grey')
plt.ylabel('Average Temperature', fontsize=18,color='Grey')
plt.title('Temp. Distribution', fontsize=20,color='Grey')
#plt.set_xticklabels(years, rotation=5)
#ax.set_title('Average global temperature including uncertainty')
plt.legend()  
plt.show()
```

Is the temperrature increasing? 
Most Definitely! Yes


```python
minyearcountry=countrytemp['Year'].min()
maxyearcountry=countrytemp['Year'].max()
maxyearcountry-minyearcountry
```

```python
country_avg = countrytemp.groupby(['Country', 'Year'])['AverageTemperature'].mean().unstack()
```

```python
country_avg.iterrows()
```

```python
for Country, row in country_avg.iterrows():
       diff=country_avg.diff(axis=1)          
```

```python
for Year in diff:
    if Year<=1999:
        diff=diff.drop([Year],axis=1)        
```

```python
for Year in range(2000,2014):
    diff[Year]=(diff[Year]/country_avg[Year-1])*100
```

```python
diff=diff.dropna()
```

```python
a=os.getcwd()
y=a+"\\Analysis\\CSVFiles\\Analysis 1"
os.chdir(y)
print(y)
```

```python
diff.to_csv('GrowthRate.csv')
```
