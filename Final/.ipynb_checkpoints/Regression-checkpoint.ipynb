#  Finding a Relation between the Month and the Average Temperature

```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
path=r'../FinalProject/Input'
```

```python
countrytemp=pd.read_csv(path+'\\'+'GlobalLandTemperaturesByCountry.csv',parse_dates=['dt'],infer_datetime_format=True)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
countrytemp['Month'] = countrytemp.dt.apply(lambda x: x.month)
```

```python
df=countrytemp.groupby(['Country','Year','Month'])[['AverageTemperature','AverageTemperatureUncertainty']].mean().reset_index()
```

```python
df1=df[(df.Country =='Brazil')]
```

```python
df1=df1.groupby(['Country','Year','Month'])[['AverageTemperature','AverageTemperatureUncertainty']].mean().reset_index()
#df1=df1.groupby(['Month','Country'])[['AverageTemperature']].mean().reset_index()
```

```python
df1=df1[-200:]
```

```python
df1.shape #200 values, 5 variables
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.set_axis_bgcolor('lightblue')
ax1.grid()
ax2 = fig.add_subplot(2, 2, 2)
ax2.set_axis_bgcolor('lightblue')
ax2.grid()
```

```python
df1.plot(kind='scatter', x='Month', y='AverageTemperature', ax=ax1,color='#ff3535',marker="o",figsize=(16, 8))
df1.plot(kind='scatter', x='Month', y='AverageTemperatureUncertainty', ax=ax2,color='#ff7d26', figsize=(16, 8))
```
