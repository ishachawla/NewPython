Finding the 10 most frequent words in Persuasion

```python
import nltk
from nltk.corpus import gutenberg 
```

```python
gutenberg.fileids() #files in gutenberg
```

```python
total_words=gutenberg.words('austen-persuasion.txt') #list of words in persuasion
```

```python
di={}
for word in total_words: # Finding frequency of words
    if word in di:
        di[word]+=1
    else:
        di[word]=1
```

```python
print(di)
```

```python
from operator import itemgetter   #Sorting a dictionary 
from collections import OrderedDict
d = OrderedDict(sorted(di.items(), key=itemgetter(1))[-10:])#Finding characters with the most occurences
```

```python
print(d) #10 most frequent words
```
