# Finding the growth rate in temperature every year from 2000-2013

```python
import pandas as pd
import numpy as np
import datetime
import calendar
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline
from bokeh.charts import Line, show, output_notebook, defaults
from bokeh.plotting import figure, output_file, show
```

```python
path=r'../FinalProject/Input'
```

```python
countrytemp=pd.read_csv(path+'\\'+'GlobalLandTemperaturesByCountry.csv',parse_dates=['dt'],infer_datetime_format=True)
```

```python
countrytemp['Year'] = countrytemp.dt.apply(lambda x: x.year)
```

```python
countrytemp=countrytemp.groupby(['Year','Country'])[['AverageTemperature','AverageTemperatureUncertainty']].mean().reset_index()
```

```python
minyearcountry=countrytemp['Year'].min()
maxyearcountry=countrytemp['Year'].max()
maxyearcountry-minyearcountry
```

```python
df = countrytemp.groupby(['Country', 'Year'])['AverageTemperature','AverageTemperatureUncertainty'].mean().reset_index()
```

```python
df=df.dropna()
```

```python
country_avg = countrytemp.groupby(['Country', 'Year'])['AverageTemperature'].mean().unstack()
```

```python
sb.set(style="white")
sb.set_color_codes("pastel")
f, ax = plt.subplots(figsize=(10, 6))

plt.plot(df['Year'][-20:], df['AverageTemperature'][-20:], label='Average Temperature', linewidth=6,color='#22357a')
plt.scatter(df['Year'][-100:], df['AverageTemperatureUncertainty'][-100:], label='Temperature Uncertainity',marker='+', color='#1b7aaa')


plt.xlim(2000, 2013)

ax.set_ylabel('Average temperature',fontsize=18,color='Grey')
ax.set_xlabel('Year',fontsize=18,color='Grey')
ax.set_title('Temperature Distribution',fontsize=20,color='Grey')
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```

Is the temperrature increasing? 
Most Definitely! Yes


```python
country_avg.iterrows()
```

```python
for Country, row in country_avg.iterrows():
       diff=country_avg.diff(axis=1)          
```

```python
for Year in diff:
    if Year<=1999:
        diff=diff.drop([Year],axis=1)        
```

```python
for Year in range(2000,2014):
    diff[Year]=(diff[Year]/country_avg[Year-1])*100
```

```python
diff=diff.dropna()
```

```python
diff.to_csv('GrowthRate.csv')
```

```python
for Year in diff: 
    diff=diff.sort_values([Year], ascending=[False])
```
